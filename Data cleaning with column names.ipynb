```python
### select all columns except one
df.columns[df.columns != 'SALARY']
```

```python
### select all columns except one using loc
df.loc[:, df.columns != 'SALARY'].columns
```

```python
### select all columns except one using difference
df.columns.difference(['SALARY'])
```

```python
### select all columns except one using negation of isin
df.loc[:, ~df.columns.isin(['SALARY'])].columns
```

```python
### select column names that begins with a word or character
df.filter(like= 'STREET').columns
```

```python
df.filter( like= 'am').columns
```

```python
### select column names that begins with a word or character using loc
df.loc[:, df.columns.str.startswith('STREET')].columns
```

```python
df.loc[:, df.columns.str.startswith('EET')].columns
```

```python
### select column names that ends with a word or character using loc
df.loc[:, df.columns.str.endswith('NAME')].columns
```

```python
df.loc[:, df.columns.str.endswith('Name')].columns
```

```python
### select column names that ends with a word or character using filter and regex name$
df.filter(regex='ame$', axis= 1).columns
```

```python
df.loc[:,df.columns.str.endswith('ame')].columns
```
